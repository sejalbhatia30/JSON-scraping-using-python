```python
street_info=[]
parking_count=[]
loc_city=[]
loc_region=[]
loc_locality=[]
reg_ent_name=[]
reg_ent_type=[]
disp_neigh=[]
name=[]
seller_label=[]
firm_name=[]
rent=[]
brokerage=[]

for x in range(1,41):
    url  = "https://search.housing.com/api/v1/rent/index/filter?results_per_page=20&poly=f97f947ffae6408ac295%2C57d63812cf899bb9b7d2%2C94c73846f1c1457094c9%2C3c69d8421a77f8f8b611&p="+str(x)
    JSONContent = requests.get(url).json()
    for entry in JSONContent['hits']: # entry -> all the dicts in JSON['hits']
        
        a={}
        b={}
        c={}
        d={}
        e={}
        f={}
        g={}
        h={}
        i={}
        j={}
        k={}
        l={}
        m={}
        a=entry['street_info']
        street_info.append(a)
        
        b=entry['parking_count']
        parking_count.append(b)
        
        c=entry['polygons_hash']['city']['name']
        loc_city.append(c)
        
        d=entry['polygons_hash']['housing_region']['name']
        loc_region.append(d)
        
        e=entry['polygons_hash']['locality']['name']
        loc_locality.append(e)
        
        f=entry['region_entity']['name']
        reg_ent_name.append(f)
        
        g=entry['region_entity']['type']
        reg_ent_type.append(g)
        
        h=entry['display_neighbourhood']
        disp_neigh.append(h)
        
        i=entry['contact_persons_info'][0]['name']
        name.append(i)
        
        j=entry['contact_persons_info'][0]['seller_label']
        seller_label.append(j)
        
        k=entry['contact_persons_info'][0]['firm_name']
        firm_name.append(k)
        
        l=entry['contact_persons_info'][0]['rent']
        rent.append(l)
        
        m=entry['contact_persons_info'][0]['brokerage']
        brokerage.append(m)
        
        
        
        
        
        
        
    


    
len(name)    
```

```python

df = pd.DataFrame()
df['street_info']=street_info
df['parking_count']=parking_count
df['loc_city']=loc_city
df['loc_region']=loc_region
df['loc_locality']=loc_locality
df['reg_ent_name']=reg_ent_name
df['reg_ent_type']=reg_ent_type
df['disp_neighbourhood']=disp_neigh
df['contact_prsn_name']=name
df['seller_label']=seller_label
df['firm_name']=firm_name
df['rent']=rent
df['brokerage']=brokerage
```
